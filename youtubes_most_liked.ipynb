```python
import csv
import matplotlib.pyplot as plt
import datetime
import calendar
import operator
import pprint
```

```python
with open('USvideos.csv', 'r', encoding='utf-8') as datafile:
    youtube_data = csv.DictReader(datafile)
    data_list = [r for r in youtube_data]
    
    data_list = sorted(data_list, key=lambda x: int(x["views"]), reverse=True)
    # print(data_list)
    uniq_res = {}

    for i in data_list:
        if uniq_res.get(i["title"]):
            if int(i["views"]) > int(uniq_res[i["title"]]["views"]):
                uniq_res[i["title"]] = i
        else:
            uniq_res[i["title"]] = i
    data_list = sorted(uniq_res.values(), key=lambda x: int(x["views"]), reverse=True)
    top_20 = data_list[:20]
    for i in top_20:
        title, views = i["title"], i["views"]
        final_data = title, views
        title_str = '"'+title+ '",'
        view_int = views+','
        print(view_int)

    
```
